# Importing Libraries

```python
import math
import random
from collections import namedtuple, deque
from itertools import count

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# we are using GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

# Replay Memory
To store past experiences

```python
Transition = namedtuple('Transition', ('state', 'action', 'reward', 'next_state', 'done'))

class ReplayMemory(object):
  def __init__(self, capacity):
    self.memory = deque([], maxlen=capacity)

  def push(self, *args):
    self.memory.append(Transition(*args))

  def sample(self, batch_size):
    return random.sample(self.memory, batch_size)

  def __len__(self):
    return len(self.memory)
```

# Making Model using DQN Algo

```python
class RLModel(nn.Module):
   def __init__(self, observations, actions):
       super(RLModel, self).__init__()
       self.layer1 = nn.Linear(observations, 128)
       self.layer2 = nn.Linear(128, 128)
       self.layer3 = nn.Linear(128, actions)

   def forward(self, x):
    x = F.relu(self.layer1(x))
    x = F.relu(self.layer2(x))
    return self.layer3(x)
```

# DQN Agent

```python
 class Agent():
    def __init__(self, state_dim, action_dim, gamma=0.95, lr=1e-3):
        self.dq_network = RLModel(state_dim, action_dim)
        self.target_network =RLModel(state_dim, action_dim)
        self.optimizer = optim.Adam(self.dq_network.parameters(), lr=lr)
        self.gamma = gamma
        self.replay_memory = ReplayMemory(1000)
        self.epsilon = 1.0
        self.epsilon_decay = 0.995
        self.epsilon_min = 0.05

    def select_action(self, state):
        if random.random() < self.epsilon:
            return random.randint(0, 3)  # Random action
        else:
            state = torch.tensor(state, dtype=torch.float32).unsqueeze(0)
            dq_values = self.dq_network(state)
            return torch.argmax(dq_values).item()

    def train(self, batch_size):
        if len(self.replay_memory) < batch_size:
            return

        batch = self.replay_memory.sample(batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)

        states = torch.tensor(states, dtype=torch.float32)
        actions = torch.tensor(actions, dtype=torch.long)
        rewards = torch.tensor(rewards, dtype=torch.float32)
        next_states = torch.tensor(next_states, dtype=torch.float32)
        dones = torch.tensor(dones, dtype=torch.float32)

        dq_values = self.dq_network(states).gather(1, actions.unsqueeze(1)).squeeze(1)
        next_dq_values = self.target_network(next_states).max(1)[0]
        target_dq_values = rewards + self.gamma * next_dq_values * (1 - dones)

        loss = nn.MSELoss()(dq_values, target_dq_values.detach())
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

    def update_target_network(self):
        self.target_network.load_state_dict(self.dq_network.state_dict())
```

# Training the Loop

```python
state_dim = envr.observation_space.shape[0]
action_dim = envr.action_space.n
agent = Agent(state_dim, action_dim)
episodes = 500
batch_size = 32

for episode in range(episodes):
    state = envr.reset()
    total_reward = 0

    while True:
        action = agent.select_action(state)
        next_state, reward, done, _ = envr.step(action)
        agent.replay_memory.push(state, action, reward, next_state, done)
        state = next_state
        total_reward += reward

        agent.train(batch_size)

        if done:
            break

        agent.update_target_network()
        agent.epsilon = max(agent.epsilon * agent.epsilon_decay, agent.epsilon_min)

        print(f"Episode: {episode + 1 }, Total Reward: {total_reward}")
```
